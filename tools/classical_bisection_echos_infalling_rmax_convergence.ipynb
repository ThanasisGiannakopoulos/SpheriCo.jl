```julia
# find proper time at r=0
# coarse
dt_c = t_list_c[2]-t_list_c[1];
proper_time_list_c = zeros(length(t_list_c));
for i in 2:length(proper_time_list_c)
    proper_time_list_c[i] = proper_time_list_c[i-1] + α_r0_t_c[i]*dt_c
end

# medium
dt_m = t_list_m[2]-t_list_m[1];
proper_time_list_m = zeros(length(t_list_m));
for i in 2:length(proper_time_list_m)
    proper_time_list_m[i] = proper_time_list_m[i-1] + α_r0_t_m[i]*dt_m
end

# fine
dt_f = t_list_f[2]-t_list_f[1];
proper_time_list_f = zeros(length(t_list_f));
for i in 2:length(proper_time_list_f)
    proper_time_list_f[i] = proper_time_list_f[i-1] + α_r0_t_f[i]*dt_f
end

#println(maximum(abs.(proper_time_list_f .-  proper_time_list_m[1:546])))
#println(maximum(abs.(proper_time_list_f .-  proper_time_list_c[1:546])))
```

```julia
# find accumulation time; we need to pairs of consequtive zero crossings

# coarse
test_sign = sign(Φ_r0_t_c[1])
zero_crossings_c = []
for i in 1:length(proper_time_list_c)
    if test_sign != sign(Φ_r0_t_c[i])
        append!(zero_crossings_c, proper_time_list_c[i])
        test_sign = sign(Φ_r0_t_c[i])
    end
end
println(zero_crossings_c)

# medium
test_sign = sign(Φ_r0_t_m[1])
zero_crossings_m = []
for i in 1:length(proper_time_list_m)
    if test_sign != sign(Φ_r0_t_m[i])
        append!(zero_crossings_m, proper_time_list_m[i])
        test_sign = sign(Φ_r0_t_m[i])
    end
end
println(zero_crossings_m)

# fine
test_sign = sign(Φ_r0_t_f[1])
zero_crossings_f = []
for i in 1:length(proper_time_list_f)
    if test_sign != sign(Φ_r0_t_f[i])
        append!(zero_crossings_f, proper_time_list_f[i])
        test_sign = sign(Φ_r0_t_f[i])
    end
end
println(zero_crossings_f)
```

```julia
zero_crossings = zero_crossings_m
proper_time_list = proper_time_list_m

println(zero_crossings)
# choose the pairs of consequtive zero crossings
un = zero_crossings[1]
unplus1 = zero_crossings[2]
um = zero_crossings[3]
umplus1 = zero_crossings[4]
# evaluate the formula for accumulation time
accum_time = (un*umplus1 - unplus1*um)/(un - unplus1 - um + umplus1)
println(accum_time)
#println(proper_time_list)
# find the similarity time T = -ln(u*-u), where u is proper time at r=0 and u* accumulationtime
similarity_time = []
i = 1
while proper_time_list[i] < accum_time
    T = -log(accum_time - proper_time_list[i])
    append!(similarity_time, T)
    i += 1
end
#println(similarity_time)
```

```julia
function cmf_conv(i::Int, ri_min, ri_max)
    it_c = its_c[i]
    it_m = its_m[i]
    it_f = its_f[i] 
    it_str_c  = lpad(it_c, 4, "0")
    it_str_m  = lpad(it_m, 4, "0")
    it_str_f  = lpad(it_f, 4, "0")
    
    Φc  =  h5read(your_dir*"/data_$((Nr-3)*2^n + 3)/data_$(it_str_c).h5","Φ")
    Πc  =  h5read(your_dir*"/data_$((Nr-3)*2^n + 3)/data_$(it_str_c).h5","Π")
    Ψc  =  h5read(your_dir*"/data_$((Nr-3)*2^n + 3)/data_$(it_str_c).h5","Ψ")
    Ac  =  h5read(your_dir*"/data_$((Nr-3)*2^n + 3)/data_$(it_str_c).h5","A")
    Bc  =  h5read(your_dir*"/data_$((Nr-3)*2^n + 3)/data_$(it_str_c).h5","B")
    DBc  =  h5read(your_dir*"/data_$((Nr-3)*2^n + 3)/data_$(it_str_c).h5","DB")
    Utldc  =  h5read(your_dir*"/data_$((Nr-3)*2^n + 3)/data_$(it_str_c).h5","Utld")
    Kc  =  h5read(your_dir*"/data_$((Nr-3)*2^n + 3)/data_$(it_str_c).h5","K")
    KBc  =  h5read(your_dir*"/data_$((Nr-3)*2^n + 3)/data_$(it_str_c).h5","KB")
    λc  =  h5read(your_dir*"/data_$((Nr-3)*2^n + 3)/data_$(it_str_c).h5","λ")
    αc  =  h5read(your_dir*"/data_$((Nr-3)*2^n + 3)/data_$(it_str_c).h5","α")
    Dαc  =  h5read(your_dir*"/data_$((Nr-3)*2^n + 3)/data_$(it_str_c).h5","Dα")
    tc   =  h5readattr(your_dir*"/data_$((Nr-3)*2^n + 3)/data_$(it_str_c).h5", "/")["time"]
    r_c  =  h5read(your_dir*"/data_$((Nr-3)*2^n + 3)/data_$(it_str_c).h5","r")

    Φm  =  h5read(your_dir*"/data_$((Nr-3)*2^(n+1) + 3)/data_$(it_str_m).h5","Φ")
    Πm  =  h5read(your_dir*"/data_$((Nr-3)*2^(n+1) + 3)/data_$(it_str_m).h5","Π")
    Ψm  =  h5read(your_dir*"/data_$((Nr-3)*2^(n+1) + 3)/data_$(it_str_m).h5","Ψ")
    Am  =  h5read(your_dir*"/data_$((Nr-3)*2^(n+1) + 3)/data_$(it_str_m).h5","A")
    Bm  =  h5read(your_dir*"/data_$((Nr-3)*2^(n+1) + 3)/data_$(it_str_m).h5","B")
    DBm  =  h5read(your_dir*"/data_$((Nr-3)*2^(n+1) + 3)/data_$(it_str_m).h5","DB")
    Utldm  =  h5read(your_dir*"/data_$((Nr-3)*2^(n+1) + 3)/data_$(it_str_m).h5","Utld")
    Km  =  h5read(your_dir*"/data_$((Nr-3)*2^(n+1) + 3)/data_$(it_str_m).h5","K")
    KBm  =  h5read(your_dir*"/data_$((Nr-3)*2^(n+1) + 3)/data_$(it_str_m).h5","KB")
    λm  =  h5read(your_dir*"/data_$((Nr-3)*2^(n+1) + 3)/data_$(it_str_m).h5","λ")
    αm  =  h5read(your_dir*"/data_$((Nr-3)*2^(n+1) + 3)/data_$(it_str_m).h5","α")
    Dαm  =  h5read(your_dir*"/data_$((Nr-3)*2^(n+1) + 3)/data_$(it_str_m).h5","Dα")
    tm   =  h5readattr(your_dir*"/data_$((Nr-3)*2^(n+1) + 3)/data_$(it_str_m).h5", "/")["time"]
    r_m  =  h5read(your_dir*"/data_$((Nr-3)*2^(n+1) + 3)/data_$(it_str_m).h5","r")

    Φf  =  h5read(your_dir*"/data_$((Nr-3)*2^(n+2) + 3)/data_$(it_str_f).h5","Φ")
    Πf  =  h5read(your_dir*"/data_$((Nr-3)*2^(n+2) + 3)/data_$(it_str_f).h5","Π")
    Ψf  =  h5read(your_dir*"/data_$((Nr-3)*2^(n+2) + 3)/data_$(it_str_f).h5","Ψ")
    Af  =  h5read(your_dir*"/data_$((Nr-3)*2^(n+2) + 3)/data_$(it_str_f).h5","A")
    Bf  =  h5read(your_dir*"/data_$((Nr-3)*2^(n+2) + 3)/data_$(it_str_f).h5","B")
    DBf  =  h5read(your_dir*"/data_$((Nr-3)*2^(n+2) + 3)/data_$(it_str_f).h5","DB")
    Utldf  =  h5read(your_dir*"/data_$((Nr-3)*2^(n+2) + 3)/data_$(it_str_f).h5","Utld")
    Kf  =  h5read(your_dir*"/data_$((Nr-3)*2^(n+2) + 3)/data_$(it_str_f).h5","K")
    KBf  =  h5read(your_dir*"/data_$((Nr-3)*2^(n+2) + 3)/data_$(it_str_f).h5","KB")
    λf  =  h5read(your_dir*"/data_$((Nr-3)*2^(n+2) + 3)/data_$(it_str_f).h5","λ")
    αf  =  h5read(your_dir*"/data_$((Nr-3)*2^(n+2) + 3)/data_$(it_str_f).h5","α")
    Dαf  =  h5read(your_dir*"/data_$((Nr-3)*2^(n+2) + 3)/data_$(it_str_f).h5","Dα")
    tf   =  h5readattr(your_dir*"/data_$((Nr-3)*2^(n+2) + 3)/data_$(it_str_f).h5", "/")["time"]
    r_f  =  h5read(your_dir*"/data_$((Nr-3)*2^(n+2) + 3)/data_$(it_str_f).h5","r")

    AH = find_AH(r_f, Bf, Af, KBf)
        
    Φcm = Φc[3:end] - Φm[3:2:end]
    Φmf = Φm[3:2:end] - Φf[3:4:end]
    
    αcm = αc[3:end] - αm[3:2:end]
    αmf = αm[3:2:end] - αf[3:4:end]

    Hc, Pc  = constraints_sbp21(Φc, Πc, Ψc, Ac, Bc, DBc, Utldc, Kc, KBc, λc, αc, Dαc, r_c);
    Hm, Pm  = constraints_sbp21(Φm, Πm, Ψm, Am, Bm, DBm, Utldm, Km, KBm, λm, αm, Dαm, r_m);
    Hf, Pf  = constraints_sbp21(Φf, Πf, Ψf, Af, Bf, DBf, Utldf, Kf, KBf, λf, αf, Dαf, r_f);
    
    Hcm = Hc[3:end] - Hm[3:2:end]
    Hmf = Hm[3:2:end] - Hf[3:4:end]
        
    Pcm = Pc[3:end] - Pm[3:2:end]
    Pmf = Pm[3:2:end] - Pf[3:4:end]
    
    plot([AH], seriestype="vline", label=L"AH", linewidth=1, color = "black", frame = true)
    plot!(r[3:ri_max], real(Φcm[1:ri_max-2]), title = "t=$(tc)",  label=L"Φ_{cm}(r)", linewidth = 2.5,
        ylim = (1.1*minimum(Φcm[1:ri_max-2]), 1.1*maximum(Φcm[1:ri_max-2])),
        color = "royalblue", frame = true)
    p0 = plot!(r[3:ri_max], 4*real(Φmf[1:ri_max-2]), title = "t=$(tc)",  label=L"4\, Φ_{mf}(r)", linewidth = 3, 
        ylim =(1.1*minimum(4*real(Φmf[1:ri_max-2])), 1.1*maximum(4*real(Φmf[1:ri_max-2]))),
        style = :dash, color = "coral", frame = true)
        
    plot([AH], seriestype="vline", label=L"AH", linewidth=1, color = "black")
    plot!(r[3:ri_max], abs.(real(Hcm[1:ri_max-2])), label=L"|H_{cm}(r)|", linewidth=2.5, 
        ylim = (0, 1.1*maximum(abs.(real(Hcm[1:ri_max-2])))),
        color = "royalblue")
    p1 = plot!(r[3:ri_max], 4*abs.(real(Hmf[1:ri_max-2])), label=L"4\, |H_{mf}(r)|", linewidth=3.0,
            ylim = (0, 1.1*maximum(4*abs.(real(Hmf[1:ri_max-2])))),
        style = :dash, color = "coral", frame = true)
        
    plot([AH], seriestype="vline", label=L"AH", linewidth=1, color = "black") 
    plot!(r[3:ri_max], abs.(real(Pcm[1:ri_max-2])), label=L"|P_{cm}(r)|", linewidth=2.5,
        ylim = (0, 1.1*maximum(abs.(real(Pcm[1:ri_max-2])))),
        color = "royalblue")
    p2 = plot!(r[3:ri_max], 4*abs.(real(Pmf[1:ri_max-2])), label=L"4\, |P_{mf}(r)|", linewidth=3.0,
        ylim =(0, 1.1*maximum(4*abs.(real(Pmf[1:ri_max-2])))),
        style = :dash, color = "coral", frame = true)
    
    plot([AH], seriestype="vline", label=L"AH", linewidth=1, color = "black") 
    plot!(r[3:ri_max], real(αcm[1:ri_max-2]), label=L"α_{cm}(r)", linewidth=2.5, 
        ylim =(1.1*minimum(real(αcm[1:ri_max-2])), 1.1*maximum(real(αcm[1:ri_max-2]))),
        color = "royalblue")
    p3 = plot!(r[3:ri_max], 4*real(αmf[1:ri_max-2]), label=L"4\, α_{mf}(r)", linewidth=3.0,
        ylim =(1.1*minimum(4*real(αmf[1:ri_max-2])),1.1*maximum(4*real(αmf[1:ri_max-2]))),
        style = :dash, color = "coral", frame = true, legend =:topright)
        
    plt = plot(p0, p1, p3, p2, layout = grid(2, 2, heights=[0.5 ,0.5, 0.5, 0.5]))
end
```
